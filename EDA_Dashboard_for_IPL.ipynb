```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# Load the dataset
df = pd.read_csv("/content/IPL_matches.csv")

# Clean column names
df.columns = df.columns.str.strip().str.lower()

# ================================
# 1. Top 10 Winning Teams
# ================================
top_winners = df['winner'].value_counts().head(10).reset_index()
top_winners.columns = ['team', 'wins']

fig1 = px.bar(top_winners, x='team', y='wins',
              title='Top 10 Winning Teams in IPL',
              color='wins', color_continuous_scale='Viridis')
fig1.show()

# ================================
# 2. Toss Decision Distribution
# ================================
fig2 = px.pie(df, names='toss_decision',
              title='Toss Decision Distribution')
fig2.show()

# ================================
# 3. Match Result Distribution
# ================================
result_counts = df['result'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(result_counts, labels=result_counts.index,
        autopct='%1.1f%%', startangle=140)
plt.title('Match Result Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()

# ================================
# 4. Top 10 Venues by Match Count
# ================================
top_venues = df['venue'].value_counts().head(10)

plt.figure(figsize=(10, 5))
top_venues.plot(kind='bar', color='orange')
plt.title('Top 10 Venues by Match Count')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
